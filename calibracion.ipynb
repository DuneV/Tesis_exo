```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
plt.style.use('classic')
```

```python
data = pd.read_excel('calibraci√≥n.xlsx')
data.head(3)
```

```python
# Plotear las columnas usando matplotlib
plt.plot(data['Aceleracion_X0'], label='$Aceleracion_{X0}$')
plt.plot(data['Aceleracion_Y0'], label='$Aceleracion_{Y0}$')
plt.plot(data['Aceleracion_X1'], label='$Aceleracion_{X1}$')
plt.plot(data['Aceleracion_Y1'], label='$Aceleracion_{Y1}$')
# Agregar una l√≠nea de referencia horizontal en y=3
plt.axhline(y=0, color='r', linestyle='--', label='$Referencia$')

# A√±adir etiquetas y leyenda
plt.title('$Aceleraciones\; vs \;muestra$')
plt.xlabel('$Muestra No.$')
plt.grid()
plt.ylabel(r'$m/s^2$')
plt.legend()

# Mostrar el gr√°fico
plt.show()
```

```python
# Plotear las columnas usando matplotlib
plt.plot(data['Aceleracion_Z0'], label='$Aceleracion_{Z0}$')
plt.plot(data['Aceleracion_Z1'], label='$Aceleracion_{Z1}$')
# Agregar una l√≠nea de referencia horizontal en y=3
plt.axhline(y=9.8, color='r', linestyle='--', label='$gravedad$')

# A√±adir etiquetas y leyenda
plt.title('Aceleraciones vs muestra')
plt.xlabel('Muestra No.')
plt.grid()
plt.ylabel(r'$m/s^2$')
plt.legend()

# Mostrar el gr√°fico
plt.show()
```
